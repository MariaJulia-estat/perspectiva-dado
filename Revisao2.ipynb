```python
import numpy as np
```

```python
vetor = np.random.randint(0, 101, 20)
print(vetor)
```

```python
#Exercício 1
media = np.mean(vetor)

desvio = np.std(vetor)

maximo = np.max(vetor)

print("media:", media)
print("desvio-padrão:", desvio)
print("o maior numero é:", maximo)
```

```python
#Exercício 2

matriz = np.random.randn(5, 5)
print(matriz)


soma_linhas = np.sum(matriz, axis=1)
print("Soma de cada linha é:" , soma_linhas)

soma_colunas = np.sum(matriz, axis=0)
print("Soma de cada coluna é:" , soma_colunas)

diagonal = np.diag(matriz)
print("A diagonal é:" , diagonal)
```

```python
#Exercício 3
def calcular(v1, v2):
    produto = np.dot(v1, v2)

    diferenca = v1 - v2
    distancia = np.linalg.norm(diferenca)

    return produto, distancia

v1 = np.array([1,2,3])
v2 = np.array([4,5,6])

produto, distancia = calcular(v1,v2)

print(produto)
print(distancia)
```

```python
#Exercício 4
#matriz 4x4 e numeros inteiros entre -5 e 5
A = np.random.randint(-5,5,(4,4))
print(A)

det = np.linalg.det(A)
print("Determinante:", det)

#Invertivel?
if det != 0:
  print("A é invertivel")

  A_inv = np.linalg.inv(A)

  # A * A^-1 = ident
  identidade = np.dot(A, A_inv)

  print("A * A^-1:")
  print(identidade)

else:
   print("A não é invertível")
```

```python
#Exercício 5
M = np.random.randn(5, 5)

# Criando matriz simétrica
B = (1/2) * (M + M.T)

print(B)

valores, vetores = np.linalg.eig(B)

print("Autovalores:")
print(valores)

print("Autovetores:")
print(vetores)

# Bv = lambda v ? verificar
v = vetores[:, 0]
lambda_ = valores[0]

print("Bv =", np.dot(B, v)) #produto escalar
print("lambda*v =", lambda_ * v)
```

```python
#Exercício 6

# Matriz A 3x3
A = np.random.randint(1, 10, (3, 3))

# Vetor
y = np.random.randint(1, 10, 3)

print(A)
print(y)

# Resolvendo sistema
x = np.linalg.solve(A, y) #resolve ax=y

print("Solução x:",x)

# ||Ax - y||
erro = np.linalg.norm(np.dot(A, x) - y) #calcula Ax e dps -y

print("Erro:", erro)
```

```python
#Exercício 7
dados = np.random.randn(100, 2)

S = np.cov(dados, rowvar=False)

print("Matriz de covariancia:", S)

# Autovalores e autovetores
valores, vetores = np.linalg.eig(S)

print("Autovalores:", valores)

print("Autovetores:", vetores)

# Direção
indice = np.argmax(valores)

maior_valor = valores[indice]
principal_direcao = vetores[:, indice]

print("Maior autovalor:", maior_valor)
print("Direção principal:", principal_direcao)
```
