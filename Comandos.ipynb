* Terminal

* Repositorio
  
1. Colocar o repositorio na pasta do computador:
   - Copiar código "code" do git hub
   - git clone + código copiado
  
2. Verificar se houve alteração:
    - Ir na pasta do repositorio, clicar com o botão direito e "abrir terminal"
    - git status: quais arquivos foram modificados, adicionados ou ainda não estão no Git
       - exemplo: arquivos modificados, arquivos prontos para commit, arquivos que deletou da pasta e ainda não atualizou o git.
    - git diff: mostra as diferenças entre as alterações feitas nos arquivos e a última versão salva no commit.
       - exemplo: linhas removidas ou adicionadas, comparação da versão atual e anterior
     
3. .
    - git add . : adiciona todas as alterações (arquivos novos e modificados) da pasta atual para o terminal, ela prepara tudo pro commit
       - adiciona todos os arquivos modificados, adiciona arquivos novos, prepara tudo para o git commit

    - git commit -m "mensagem": salva as alterações que estão no terminal no histórico do Git, criando uma mensagem
        - cria um novo commit, registra as mudanças no histórico, associa uma mensagem explicando o que foi feito
  
    - git push: envia os commits do seu repositório local para o repositório do github
        - envia as alterações para o repositório remoto, atualiza o projeto online, sincroniza o repositório local com o github
     
    - Faz os passos do item 2 antes de salvar (git push) para verificar se foi atualizado
  
    - git log: lista de commits, informações do autor, data e mensagem de cada commit

4..
* ls: após abrir o terminal da pasta, escreve o ls pra mostrar todas as pastas que tem no reposítorio
* cd + nome da pasta: entra na pasta escolhida, muda o diretório atual no terminal
  - cd .. sai da pasta atual, retorna para o anterior do diretório

* Atualizar o arquivo
  - Para modificar cada arquivo por vez: git add + nome do arquivo
  - git commit -m "mensagem"

* Antes de dar o commit, é possível retirar a modificação do git add
  - git restore -- staged + nome do arquivo
 
* Desfazer alteração que não está no git add
  - git restore + nome do arquivo
 
* git pull: sincroniza

* cat + nome do arquivo: mostra o conteúdo do arquivo
