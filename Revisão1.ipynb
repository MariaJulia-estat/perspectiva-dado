```python
#Passo 2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Passo 1

distribuicoes = {
    "Normal": {"params": {"loc": 0, "scale": 1}, "dados": None, "estatisticas": None},
    "Exponencial": {"params": {"scale": 2}, "dados": None, "estatisticas": None},
    "Uniforme": {"params": {"low": -1, "high": 1}, "dados": None, "estatisticas": None}
}

pd.DataFrame(distribuicoes)
```

```python
#Passo 3
x = np.random.normal(0,1,1000)
y = np.random.uniform(-1,1,1000)
z = np.random.exponential(2,1000)

distribuicoes['Normal']["dados"] = x
distribuicoes['Uniforme']["dados"] = y
distribuicoes['Exponencial']["dados"] = z
```

```python
#Passo 4

media_x = np.mean(x)
media_y = np.mean(y)
media_z = np.mean(z)

mediana_x = np.median(x)
mediana_y = np.median(y)
mediana_z = np.median(z)

var_x = np.var(x)
var_y = np.var(y)
var_z = np.var(z)

dp_x = np.std(x)
dp_y = np.std(y)
dp_z = np.std(z)

est_x = {"media":media_x,"mediana":mediana_x,"variancia":var_x,"dp":dp_x}
est_y = {"media":media_y,"mediana":mediana_y,"variancia":var_y,"dp":dp_y}
est_z = {"media":media_z,"mediana":mediana_z,"variancia":var_z,"dp":dp_z}

distribuicoes['Normal']["estatisticas"] = est_x
distribuicoes['Uniforme']["estatisticas"] = est_y
distribuicoes['Exponencial']["estatisticas"] = est_z
```

```python
#Passo 5
plt.subplot(1,3,1)
plt.hist(x)
plt.title("Distribuição Normal")
plt.show()

plt.subplot(1,3,2)
plt.hist(y)
plt.title("Distribuição Uniforme")
plt.show()

plt.subplot(1,3,3)
plt.hist(z)
plt.title("Distribuição Exponencial")
plt.show()
```

```python
#Passo 6
df_estatisticas = pd.DataFrame({
    "Normal": est_x,
    "Uniforme": est_y,
    "Exponencial": est_z
})

pd.DataFrame(df_estatisticas)
```

```python
#Passo 7

distribuicoes['Poisson'] = {"params": {"lam": 3}, "dados": None, "estatisticas": None}
pd.DataFrame(distribuicoes)
```

```python
w = np.random.poisson(3,1000)
distribuicoes['Poisson']["dados"] = w
```

```python
media_w = np.mean(w)
mediana_w = np.median(w)
var_w = np.var(w)
dp_w = np.std(w)
est_w = {"media":media_w,"mediana":mediana_w,"variancia":var_w,"dp":dp_w}

distribuicoes['Poisson']["estatisticas"] = est_w
```

```python
df_estatisticas = pd.DataFrame({
    "Normal": est_x,
    "Uniforme": est_y,
    "Exponencial": est_z,
    "Poisson": est_w
})

pd.DataFrame(df_estatisticas)
```

```python
#Passo 8
min_x, max_x = np.min(x), np.max(x)
distribuicoes['Normal']['estatisticas'].update({"min": min_x, "max": max_x, "q1": np.percentile(x, 25), "q2": np.percentile(x, 50), "q3": np.percentile(x, 75)})


min_y, max_y = np.min(y), np.max(y)
distribuicoes['Uniforme']['estatisticas'].update({"min": min_y, "max": max_y, "q1": np.percentile(y, 25), "q2": np.percentile(y, 50), "q3": np.percentile(y, 75)})


min_z, max_z = np.min(x), np.max(z)
distribuicoes['Exponencial']['estatisticas'].update({"min": min_z, "max": max_z, "q1": np.percentile(z, 25), "q2": np.percentile(z, 50), "q3": np.percentile(z, 75)})


min_w, max_w = np.min(w), np.max(w)
distribuicoes['Poisson']['estatisticas'].update({"min": min_w, "max": max_w, "q1": np.percentile(w, 25), "q2": np.percentile(w, 50), "q3": np.percentile(w, 75)})

```

```python
df_estatisticas = pd.DataFrame({
    "Normal": est_x,
    "Uniforme": est_y,
    "Exponencial": est_z,
    "Poisson": est_w
})

pd.DataFrame(df_estatisticas)
```

```python
#Passo 9
plt.boxplot([x, y, z, w], labels=["Normal", "Uniforme", "Exponencial", "Poisson"],widths=0.6)

plt.title("boxplot das distribuiçoes")
plt.ylabel("valores")

plt.show()
```

```python
#Passo 10

# Normal
mtn = distribuicoes['Normal']['params']['loc']
distribuicoes['Normal']['estatisticas']['media_teorica'] = mtn

# Uniforme
low = distribuicoes['Uniforme']['params']['low']
high = distribuicoes['Uniforme']['params']['high']
mtu = (low + high) / 2
distribuicoes['Uniforme']['estatisticas']['media_teorica'] = mtu

# Exponencial
mte = distribuicoes['Exponencial']['params']['scale']
distribuicoes['Exponencial']['estatisticas']['media_teorica'] = mte

# Poisson
mtp = distribuicoes['Poisson']['params']['lam']
distribuicoes['Poisson']['estatisticas']['media_teorica'] = mtp
```

```python
df_stats = pd.DataFrame({
    "Normal": distribuicoes['Normal']['estatisticas'],
    "Uniforme": distribuicoes['Uniforme']['estatisticas'],
    "Exponencial": distribuicoes['Exponencial']['estatisticas'],
    "Poisson": distribuicoes['Poisson']['estatisticas']
})

pd.DataFrame(df_stats)
```

```python
nomes = ['Normal',  'Uniforme', 'Exponencial', 'Poisson']

media_amostral = [
    distribuicoes['Normal']['estatisticas']['media'],
    distribuicoes['Uniforme']['estatisticas']['media'],
    distribuicoes['Exponencial']['estatisticas']['media'],
    distribuicoes['Poisson']['estatisticas']['media']
]

media_teorica = [
    distribuicoes['Normal']['estatisticas']['media_teorica'],
    distribuicoes['Uniforme']['estatisticas']['media_teorica'],
    distribuicoes['Exponencial']['estatisticas']['media_teorica'],
    distribuicoes['Poisson']['estatisticas']['media_teorica']
]

x_posição = np.arange(len(nomes))

plt.bar(x_posição, media_amostral,width=0.4, label='Amostral')
plt.bar(x_posição + 0.4, media_teorica, width=0.4, label='Teorica')

plt.xticks(x_posição, nomes)
plt.title("Media Amostral vs Teorica")
plt.ylabel("Valor da Media")
plt.legend()

plt.show()
```
