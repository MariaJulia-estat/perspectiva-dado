## Arma do crime

#### **`requests`**
- Biblioteca utilizada para **fazer requisições HTTP**.
- Permite acessar páginas web e obter o conteúdo delas, como HTML, JSON ou arquivos de texto.
- Fundamental para capturar o código-fonte de uma página para análise.

#### **`BeautifulSoup`** (do pacote `bs4`)
- Biblioteca usada para **analisar e manipular documentos HTML ou XML**.
- Facilita a extração de informações específicas, como títulos, links ou tabelas, a partir do conteúdo HTML obtido.

### Fluxo básico:
1. Utilize `requests` para fazer a requisição e capturar o conteúdo HTML da página.
2. Use `BeautifulSoup` para analisar o HTML e extrair os dados que deseja.

````python
import json, re

# Extraindo título e descrição (esses funcionam normal)
title_tag = soup.find("meta", property="og:title")
title = title_tag["content"] if title_tag else "Sem título"

desc_tag = soup.find("meta", attrs={"name": "description"})
description = desc_tag["content"] if desc_tag else "Sem descrição"

# Extraindo data, categoria e tags do JSON-LD
json_ld_tag = soup.find("script", attrs={"type": "application/ld+json"},
                        string=re.compile("datePublished"))
if json_ld_tag:
    data_json = json.loads(json_ld_tag.string)
    date = data_json.get("datePublished", "Sem data")
    section = data_json.get("articleSection", "Sem categoria")
else:
    date = "Sem data"
    section = "Sem categoria"

# Extraindo tags do dataLayer
datalayer_tag = soup.find("script", string=re.compile("publication_date"))
tags = []
if datalayer_tag:
    tags_match = re.search(r'"tags":\[(.*?)\]', datalayer_tag.string)
    if tags_match:
        tags = [t.strip('"') for t in tags_match.group(1).split(',')]

# Extraindo texto
paragrafos = soup.find_all("p")
texto = "\n".join([p.get_text(strip=True) for p in paragrafos
                   if p.get_text(strip=True)])

print("Título:", title)
print("Descrição:", description)
print("Categoria:", section)
print("Tags:", tags)
print("Data:", date)
````

## PERDEU!!!



Agora vamos ~~roubar~~ extrair o máximo de coisas que conseguirmos sem deixar (muitas) evidências.

````python
import time
import random
````

````python
numero_paginas = 13
# numero_paginas = 5

links_noticias = []

for p in range(1, numero_paginas + 1):
    url = f"https://exame.com/noticias-sobre/olimpiadas/{p}/"
    response = requests.get(url, headers=headers)
    print(p, response)

    soup = BeautifulSoup(response.text, "html.parser")
    todos_h3 = soup.find_all("h3", class_="m-0 p-0 xl:text-pretty headline-extra-small text-colors-text")
    links = [h3.find('a', href=True)["href"] for h3 in todos_h3
             if h3.find('a', href=True)]
    links_noticias += links

    tempo_espera = random.uniform(0, 0.01)
    time.sleep(tempo_espera)
````

````python
len(links_noticias)
````

````python
import json
import pathlib

data_dir = pathlib.Path("data")
data_dir.mkdir(exist_ok=True)

for i, link in enumerate(links_noticias):
    full_url = link  # link já vem completo na Exame

    response = requests.get(full_url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    title_tag = soup.find("meta", property="og:title")
    title = title_tag["content"] if title_tag else "Sem título"

    desc_tag = soup.find("meta", attrs={"name": "description"})
    description = desc_tag["content"] if desc_tag else "Sem descrição"

    section_tag = soup.find("meta", property="article:section")
    section = section_tag["content"] if section_tag else "Sem categoria"

    date_tag = soup.find("meta", property="article:modified_time")
    date = date_tag["content"] if date_tag else "Sem data"

    paragrafos = soup.find_all("p")
    texto = "\n".join([p.get_text(strip=True) for p in paragrafos
                       if p.get_text(strip=True)])

    informacoes = {
        "url": full_url,
        "titulo": title,
        "descricao": description,
        "categoria": section,
        "texto": texto,
        "data": date,
    }

    filename = f"noticia_{i}.json"
    file_path = data_dir / filename

    with file_path.open("w", encoding="utf-8") as f:
        json.dump(informacoes, f, ensure_ascii=False, indent=4)

    print(f"Arquivo salvo: {file_path}")
    tempo_espera = random.uniform(0, 0.5)
    time.sleep(tempo_espera)
````

## Sequestro relâmpago

````python
from joblib import Parallel, delayed

def processar_link(i, link):
    full_url = link  # link já vem completo na Exame

    response = requests.get(full_url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    title_tag = soup.find("meta", property="og:title")
    title = title_tag["content"] if title_tag else "Sem título"

    desc_tag = soup.find("meta", attrs={"name": "description"})
    description = desc_tag["content"] if desc_tag else "Sem descrição"

    section_tag = soup.find("meta", property="article:section")
    section = section_tag["content"] if section_tag else "Sem categoria"

    date_tag = soup.find("meta", property="article:modified_time")
    date = date_tag["content"] if date_tag else "Sem data"

    paragrafos = soup.find_all("p")
    texto = "\n".join([p.get_text(strip=True) for p in paragrafos
                       if p.get_text(strip=True)])

    informacoes = {
        "url": full_url,
        "titulo": title,
        "descricao": description,
        "categoria": section,
        "texto": texto,
        "data": date,
    }


    file_path = data_dir / f"noticia_{i}.json"
    with file_path.open("w", encoding="utf-8") as f:
        json.dump(informacoes, f, ensure_ascii=False, indent=4)

    print(f"Arquivo salvo: {file_path}")

    tempo_espera = random.uniform(0, .5)
    time.sleep(tempo_espera)

# Executa em paralelo com threads
Parallel(n_jobs=8, prefer="threads")(
    delayed(processar_link)(i, link) for i, link in enumerate(links_noticias)
)
````

````python
import pandas as pd
import pathlib
import json

data_dir = pathlib.Path("data")

registros = []
for arquivo in data_dir.glob("noticia_*.json"):
    with arquivo.open("r", encoding="utf-8") as f:
        registros.append(json.load(f))

df = pd.DataFrame(registros)
print(df.shape)
df.head()
````

````python
df.to_csv("olimpiadas_exame.csv", index=False, encoding="utf-8")
print("Base salva!")
````
